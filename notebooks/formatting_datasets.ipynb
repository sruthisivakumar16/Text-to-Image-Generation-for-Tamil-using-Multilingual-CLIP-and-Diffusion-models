```python
!pip install datasets
from datasets import Dataset as HFDataset, Features, Image, Sequence, Value
import torch
```

```python
#paths for embeddings file
test_embeddings_file = '/content/drive/MyDrive/Colab Notebooks/thesis/embeddings/labse/labse_test_2016_embeddings.pth'
train_embeddings_file = '/content/drive/MyDrive/Colab Notebooks/thesis/embeddings/labse/labse_train_embeddings.pth'
images_dir = '/content/drive/MyDrive/Colab Notebooks/thesis/MMDravi/flickr30k_images/flickr30k_images'

test_embeddings_data = torch.load(test_embeddings_file)
train_embeddings_data = torch.load(train_embeddings_file)
```

```python
#define the features for the dataset according to what the training script expects
features = Features({
    "image": Image(),
    "caption": Value(dtype="string")
})

#create a dictionary with image paths and captions
test_dataset_dict_cap = {
    "image": [f"{images_dir}/{item['image_path']}" for item in test_embeddings_data],
    "caption": [item['caption'] for item in test_embeddings_data]
}

#create the Hugging Face dataset
test_dataset_cap = HFDataset.from_dict(test_dataset_dict_cap, features=features)
```

```python
features = Features({
    "image": Image(),
    "caption": Value(dtype="string")
})

# Create a dictionary with image paths and captions
train_dataset_dict_cap = {
    "image": [f"{images_dir}/{item['image_path']}" for item in train_embeddings_data],
    "caption": [item['caption'] for item in train_embeddings_data]
}

# Create the Hugging Face dataset
train_dataset_cap = HFDataset.from_dict(train_dataset_dict_cap, features=features)
```

```python
#Save the test dataset with captions
test_dataset_cap.save_to_disk('/content/drive/MyDrive/Colab Notebooks/thesis/datasets/hf_test_2016_dataset_cap')
```

```python
#save the train dataset with captions
train_dataset_cap.save_to_disk('/content/drive/MyDrive/Colab Notebooks/thesis/datasets/hf_train_dataset_cap')
```
